```python

##############################################################################

# Welcome to BanderSnatch - A Virtual Trap

##############################################################################

# Created on : Oct 19 '2020
# Author     : Aveka Patawari

"""
DocString:
    
    A) Introduction:
        The characters and the concept of this adventure game is partially 
        derived from the Movie - Black Mirror: Bandersnatch. This is an 
        imaginative version of the original movie. To win, the user needs to 
        complete 3 tasks assigned to him/her by PAX, a computer Trojan.
       
        Goal - The ultimate mission is to free Stefan, a young programmer who
        is trapped inside a video game called Bandersnatch by a Trojan named 
        PAX. The user can do this by winning all the 3 tasks set by PAX.
       
Task 1: THE ROOM - choose the correct object in the room to win this round.
Task 2: CASINO ROYALE - roll the dice and guess the outcome to win this round.
Task 3: THE ULTIMATE MAZE TRAP - guess the correct dicrection to get out
        of the maze, win the game and free Stefan.

    B) Known Bugs and/or Errors:
        None.

"""
# importing necessary packages
import random
import time
import sys



# defining function start 
def start():
    
    
    # setting global object 
    global player
    
    print("""\n\n
Welcome to BanderSnatch - A virtual Trap! 
A Game that is going to change your life!!!\n """)
    player = str (input (prompt = "Enter your name Player_1 :   "))
    player = player.capitalize()
    
    intro = ([f"""\n      ****   W A R N I N G   ****
    
\nWAIT... {player}, This is a WARNING for you! 
I am Stefan, I can't stay here for long... so listen carefully..\n"""] +
[f"""This is not just any game you are playing... {player},
I have been trapped inside this game by a Trojan named PAX, and now he is going to trap you too!
You have to win all the 3 rounds to free yourself and me!\n"""] +
["\nHELLLPPPP Meeeee.... ple... \n"])
    
    
    # using time.sleep function to print the above sentences individually
    for sentence in intro:
        print(sentence, end='')
        time.sleep(1.5)  # pause for 1.5 second for each sentence
        
    input(prompt = "\n * Press <enter> to Continue * \n")
    
    print ("""
*****

    You Suddenly see a light coming out of the screen and you are Magically
    pulled into the game. It's dark inside and you can see computer codes blinking around you.
    You look for Stefan but he is nowhere to be found. 
    
    Now, the only way out is to win all the 3 rounds. 
    
                                                   *****
                                                   \n""")
    
    input(prompt = "\n *** Press <enter> to Play *** \n") # prompting user to press enter to play
    
    room()  # calling room to start task_1
    
    
##############################################################################
# Constructing / defining room (room) for the first task
##############################################################################
    
def room():
    
    # room introduction for user
    print (f"""
{"#" * 90}
    
    ## This is round 1. You are Transported into a dark room which is locked from inside. ##
    ## You have to pick an object that will help you get out of the room.                 ##\n
    
                                 T A S K     1
    
    """)
    
    print(f"""
    PAX : Welcome to round 1. To unlock the door you must select the correct object. 
    There is a key hidden inside one of the objects.
        
        1. Photoframe
        2. Book
        3. Torch
        4. Jar """)
    
    obj = input(prompt = f"""\nSo, what do you choose {player}? """)
    obj = obj.lower()   # setting output value of obj to lower case
    
    
    # setting input value for user in obj
    if obj == '1' or obj == 'a' or obj == 'photoframe' or obj == 'pick photoframe' or obj == 'collect photoframe':
        print (f"\nPAX : Wrong choice {player}. Now, you will be trapped in this photoframe forever!\n")
        fail()         # calling fail if player loses
        
    elif obj == '2' or obj == 'b' or obj == 'book' or obj == 'pick book' or obj == 'collect book':
        print (f"""\n
        
                                       ###
                          ***** K E Y #####
                           #*          ###
         
         
        PAX : Looks like you found the key in the book to unlock the door. 
        You win round 1... Don't look so happy... the game has just begun!
               \n""")
        input(prompt = "\n *** Press <enter> to Continue*** \n")
        casino()       # calling second room if player wins
        
        
    elif obj == '3' or obj == 'c' or obj == 'torch' or obj == 'pick torch' or obj == 'collect torch':
        print (f"\nPAX : Wrong choice {player}. You lose!")
        fail()         # calling fail
               
    elif obj == '4' or obj == 'd' or obj == 'jar' or obj == 'pick jar' or obj == 'collect jar':
        print (f"PAX : Wrong choice {player}. You are trapped in this jar for eternity!")
        fail()         # calling fail
        
    else:
        print (f"""Something's wrong. Try again. (HINT : input corresponding value from object list).""")
        room()         # calling room again if user input is invalid
        
           
        
##############################################################################
# Constructing / defining room (casino) for the second task
##############################################################################

def casino():
    
    # introduction for user 
    print (f""" 
{"#" * 90}

    ## You cleared the first task but you are still trapped in this video-game. ##
    ## You are Transported to WinStar World Casino for task 2.                  ##
                            
                                    T A S K     2
    """)
    
    print (f"""
PAX : So you made it to round 2! Must be your lucky day {player}.
      Since you are here your next task is to guess a random number before the dice is rolled.
      BUT...You only have 3 chances...""")
    
    
    # declaring objects
    value = random.randint(1, 6) # number remains constant for each guess
    dice_roll = 3
    
    # looping until a user is out of guesses
    while dice_roll > 0:
        guess = input(prompt ="\nGuess a number between 1 and 6 inclusive  :  ") # prompting for user input
        
        # starting the try and except function
        try:
            guess = int(guess)    # converting guess to integer
            if guess == value: 
                print ("\nRolling the dice...")
                print("\nPAX : *** W I N N E R *** ! You got lucky...Beginner's luck maybe?\n")
                maze()            # calling room 3 when you win
                
                
            # conditional statement based on user input
            elif guess != value and guess < 7 and guess > 0:
                print ("Rolling the dice...")
                dice_roll -= 1

                # looping 3 times untill the user guesses correctly
                if dice_roll == 2:
                    print(f"""\nPAX : Ah! Looks like your guess is incorrect.
    You still have {dice_roll} chances left, but don't get your hopes high {player}.""")

                elif dice_roll == 1:
                    print(f"\nPAX : Your guess is wrong again {player}. You have {dice_roll} chance left. ")

                elif dice_roll == 0:
                    print(f"\nPAX : BwaHAHAHA...You lose {player}. You could'nt even guess the number {value}.")
                    fail()

        # setting except function for ValueError
        except ValueError:
            print(f"\nWrong Input. Please try again!")
            break
                 
    fail()
    
##############################################################################
# Constructing / defining room (maze) for the last task
##############################################################################
                
def maze():
    
    # introduction for user
    print (f"""
{"#" * 90}   
    
    ## This is the last and the 3rd round of the game. You are dropped into a maze with no directions. ##
    ## Well, to win this round you have rely on your instincts and get out of the maze safely.         ##
    
                           T H E     W I N N I N G    C H A L L E N G E 
    """)
    
    print(f"""
    PAX : You made it till here, impressive! Let's see if you can get out of this maze {player}.
    You can choose to move to any one of the 4 diections of your choose. If incorrect, you face the consequences.
    So you thought it would be easy ?... HAHAHAHA... 
    
    # It's dark inside, you see some light in the west. You then suddenly hear a sound coming from east. #
    # There are rodent like creatures moving towards you from behind. You have to make a choice fast! 

                    1. East
                    2. West
                    3. North
                    4. South
        """)
    # asking user to input directions
   

    
    path = input(prompt = f"""\nSo, Where do you wish to go {player} ? Enter a Direction: """)
    path = path.lower()  # setting user input to lower case
    
    # setting input options that can be entered by user
    
    if path == '1' or path == 'a' or path == 'east' or path == 'go east' or path == 'right' or path == 'e':
        print (f"Wrong Path {player}. You encountered a troll on this path and he swallowed you!")
        fail()           # calling fail if user loses

    elif path == '2' or path == 'b' or path == 'west' or path == 'go west' or path == 'left' or path == 'w':
        print (f"PAX : YOU WIN {player} !!! It can't be!!!!!\n")
        print ("""
        YOU ARE AMAZING {player}! You were sucessfully able to get out of the maze and free Stefan
        from captivity and by defeating PAX with your wits and luck.


                                  ***  Y O U       W I N   ***

        """)
        sys.exit(0)       # exit game when you win


    elif path == '3' or path == 'c' or path == 'north' or path == 'go north' or path == 'forward' or path == 'n':
        print (f"Your instincts were wrong. You are now caught in a never ending python loop.")
        fail()            # calling fail

    elif path == '4' or path == 'd' or path == 'south' or path == 'go south' or path == 'back' or path == 's':
        print (f"You chose the Wrong  path {player}, and it all went south. you lose!")
        fail()            # calling fail 

    else:
        print (f"""Invalid input. Try again! ( Surprise HINT : Follow the light).""")
        maze()            # calling room (maze) again if user input is invalid


        
              
    
# defining fail function
def fail():

    # prompting user to choose again to play
    replay = input(prompt = f"\nWould you like to play again {player} ? (Yes/No)  : ")
    replay = replay.lower()
    
    if replay == 'yes' or replay == 'y':
        start()    # start the game from the beginning 

    elif replay == 'no' or replay == 'n':
        print("\nThanks for playing the game!") # exit game

    else:
        print("\n*** Wrong input. Please try again *** ")
        fail()   # for invalid user input

    




##############################################################################
# START GAME
##############################################################################
start()


```
